### ✅ 2. Research Background
##### A. Health Impacts of PM2.5
* PM2.5 refers to fine particulate matter with a diameter of 2.5 micrometers or less.
* According to OECD data, as of 2020, South Korea's PM2.5 concentration stands at 25.3 µg/m³, over <span style="color:orange;">five times</span> worse than Finland, the least polluted country (4.9 µg/m³).
* WHO identifies PM2.5 as a <span style="color:orange;">major pollutant</span> linked directly to cardiovascular disease, respiratory illness, and premature death.
* *[Ref] [**How Much Can EVs Reduce Air Pollution?**](https://www.si.re.kr/node/59902?utm_source=)*

##### B. Direct Link to Vehicle Emissions
* PM2.5 is a <span style="color:orange;">primary pollutant</span> emitted by internal combustion engine vehicles (diesel). By 2050, a projected EV adoption rate of 54% is expected to reduce pollutant emissions by 77% compared to 2016 levels.
* However, while increasing EV adoption is expected to reduce PM2.5 levels, <span style="color:orange;">secondary contributors</span> such as tire and brake wear must also be considered.
* *[Ref] [**The Real Culprits of Vehicle-Related Fine Dust**](https://www.hani.co.kr/arti/science/science_general/921777.html)*

##### C. Policy Priorities
* South Korea's <span style="color:orange;">‘Seasonal Fine Dust Management Policy’</span> aims to improve air quality by targeting PM2.5.
* PM2.5, classified as ultrafine dust, has greater adverse effects on human health compared to PM10, making it a key <span style="color:orange;">indicator</span> for assessing air quality improvements.
* *[Ref] [**Lowest PM2.5 Levels Recorded Over the Past Five Years During the 5th Seasonal Management Period**](https://me.go.kr/home/web/board/read.do?pagerOffset=0&maxPageItems=10&maxIndexPages=10&searchKey=title&searchValue=%EC%A0%9C5%EC%B0%A8+%EB%AF%B8%EC%84%B8%EB%A8%BC%EC%A7%80+%EA%B3%84%EC%A0%88%EA%B4%80%EB%A6%AC%EC%A0%9C&menuId=10525&orgCd=&boardId=1667850&boardMasterId=1&boardCategoryId=&decorator=)*

### ✅ 3. Data Overview
##### A. Data Sources

* Regional EV statistics (KEPCO, https://www.data.go.kr/data/15039554/fileData.do)
    - To address timeline inconsistencies, the analysis will use December 2023 data instead of the latest July 2024 data.

##### B. Key Variables
* Regional EV ownership ratio: {(Number of EVs in a region / Total number of EVs nationwide) * 100}
* Air pollution variables: PM2.5, PM10, NO₂ (vehicle emissions-related), SO₂ (industrial emissions-related), CO (traffic congestion-related).
* Time variable: Monthly data for time-series analysis.
