### ✅ 2. 연구 배경
##### A. PM2.5 건강 영향
* PM2.5는 미세먼지의 한 종류로 지름 2.5㎛ 이하의 먼지를 말함.
* OECD 자료에 따르면, 2020년 기준 한국의 미세먼지(PM2.5)농도는 25.3㎍/㎥으로 미세먼지 오염도가 가장 낮은 핀란드(4.9㎍/㎥)에 비하면 <span style="color:orange;">5배 이상</span> 심각함.
* WHO는 PM2.5를 심혈관 질환, 호흡기 질환, 조기 사망과 직접적으로 연관된 <span style="color:orange;">주요 오염물질</span>로 규정하고 있음.
* *[Ref] [**전기차 늘면 대기오염 얼마나 줄어들까?**](https://www.si.re.kr/node/59902?utm_source=)*

##### B. 차량 배출과의 직접적 연관성
* PM2.5는 <span style="color:orange;">내연기관 차량(디젤)의 주요 배출물질</span>로 2050년까지 EV(Electric-Vechicles)보급률이 54%에 도달하면 2016년 대비 오염물질 배출량 77% 감소 예상.
* 단, EV 보급률 증가가 PM2.5 감소에 기여할 것으로 예측되지만, 타이어와 브레이크 마모와 같은 <span style="color:orange;">2차적 미세먼지 발생 원인</span> 역시 고려해야 함.
* *[Ref] [**자동차 미세먼지 배출 ‘진범’은 따로 있다**](https://www.hani.co.kr/arti/science/science_general/921777.html)*
  
##### C. 정책적 우선순위
* 국내 <span style="color:orange;">‘미세먼지 계절관리제’</span>는 PM2.5를 중심으로 한 대기질 개선을 목표로 함.
* 초미세먼지로 분류되는 PM2.5는 미세먼지(PM10)보다 인체에 미치는 영향이 더 커 실제로 대기질 개선 효과를 평가할 수 있는 <span style="color:orange;">척도</span>로 사용됨.
* *[Ref] [**제5차 미세먼지 계절관리제 기간 초미세먼지 농도 지난 5년 이래 최저**](https://me.go.kr/home/web/board/read.do?pagerOffset=0&maxPageItems=10&maxIndexPages=10&searchKey=title&searchValue=%EC%A0%9C5%EC%B0%A8+%EB%AF%B8%EC%84%B8%EB%A8%BC%EC%A7%80+%EA%B3%84%EC%A0%88%EA%B4%80%EB%A6%AC%EC%A0%9C&menuId=10525&orgCd=&boardId=1667850&boardMasterId=1&boardCategoryId=&decorator=)*
